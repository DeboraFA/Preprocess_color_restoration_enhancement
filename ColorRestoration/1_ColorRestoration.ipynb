```python
import os
import cv2
import glob
import matplotlib.pyplot as plt
from PIL import Image




from DCP import DCP
from GBdehazingRCorrection import GBdehazingRC
from IBLA import IBLA
from LowComplexityDCP import LowComplexityDCP
from MIP import MIP
from NewOpticalModel import NewOpticalModel
from UDCP import UDCP
from Rows import Rows
from ULAP import ULAP
```

```python


def color_restoration(img_path, preprocess):

    if preprocess == 'DCP':
        sceneRadiance = DCP(img_path)
    elif preprocess == 'GBdehazingRCorrection':
        sceneRadiance = GBdehazingRC(img_path)
    elif preprocess == 'IBLA':
        sceneRadiance = IBLA(img_path) 
    elif preprocess == 'LowComplexityDCP':
        sceneRadiance = LowComplexityDCP(img_path) 
    elif preprocess == 'MIP':
        sceneRadiance = MIP(img_path) 
    elif preprocess == 'NewOpticalModel':
        sceneRadiance = NewOpticalModel(img_path) 
    elif preprocess == 'Rows':
        sceneRadiance = Rows(img_path)         
    elif preprocess == 'UDCP':
        sceneRadiance = UDCP(img_path) 
    elif preprocess == 'ULAP':
        sceneRadiance = ULAP(img_path) 
                
    return sceneRadiance
    
```

```python
img_path = 'img.png'

DCP = color_restoration(img_path, 'DCP')
GBdehazingRCorrection = color_restoration(img_path, 'GBdehazingRCorrection')
IBLA = color_restoration(img_path, 'IBLA')
LowComplexityDCP = color_restoration(img_path, 'LowComplexityDCP')
MIP = color_restoration(img_path, 'MIP')
NewOpticalModel = color_restoration(img_path, 'NewOpticalModel')
UDCP = color_restoration(img_path, 'UDCP')
Rows = color_restoration(img_path, 'Rows')
ULAP = color_restoration(img_path, 'ULAP')



# display results
fig, ax = plt.subplots(nrows=3, ncols=3, figsize=(10, 8))

ax = ax.flatten()
for a in ax:
    ax[0].imshow(DCP)
    ax[0].set_title('DCP', fontsize=12)
    ax[0].set_xticks([])
    ax[0].set_yticks([])

    ax[1].imshow( cv2.cvtColor(GBdehazingRCorrection, cv2.COLOR_BGR2RGB))
    ax[1].set_title('GBdehazingRCorrection', fontsize=12)
    ax[1].set_xticks([])
    ax[1].set_yticks([])

    ax[2].imshow(IBLA)
    ax[2].set_title('IBLA', fontsize=12)
    ax[2].set_xticks([])
    ax[2].set_yticks([])
    
    ax[3].imshow(LowComplexityDCP)
    ax[3].set_title('LowComplexityDCP', fontsize=12)
    ax[3].set_xticks([])
    ax[3].set_yticks([])
    
    ax[4].imshow( cv2.cvtColor(MIP, cv2.COLOR_BGR2RGB))
    ax[4].set_title('MIP', fontsize=12)
    ax[4].set_xticks([]) 
    ax[4].set_yticks([]) 

    ax[5].imshow(NewOpticalModel)
    ax[5].set_title('NewOpticalModel', fontsize=12)
    ax[5].set_xticks([]) 
    ax[5].set_yticks([]) 

    ax[6].imshow(cv2.cvtColor(UDCP, cv2.COLOR_BGR2RGB))
    ax[6].set_title('UDCP', fontsize=12)
    ax[6].set_xticks([]) 
    ax[6].set_yticks([]) 

    ax[7].imshow(Rows)
    ax[7].set_title('Rows', fontsize=12)
    ax[7].set_xticks([])
    ax[7].set_yticks([])

    ax[8].imshow(cv2.cvtColor(ULAP, cv2.COLOR_BGR2RGB))
    ax[8].set_title('ULAP', fontsize=12)
    ax[8].set_xticks([])
    ax[8].set_yticks([])
```
