```python
import os
import cv2
import glob
import matplotlib.pyplot as plt


from CLAHE import CLAHE
from GC import GC
from HE import HE
from ICM import ICM
from Rayleigh import Rayleigh
from RGHS import RGHS
from UCM import UCM
```

```python
def image_enhancement(img_path, preprocess):

    if preprocess == 'CLAHE':
        sceneRadiance = CLAHE(img_path)
    elif preprocess == 'GC':
        sceneRadiance = GC(img_path)
    elif preprocess == 'HE':
        sceneRadiance = HE(img_path)
    elif preprocess == 'ICM':
        sceneRadiance = ICM(img_path)
    elif preprocess == 'Rayleigh':
        sceneRadiance = Rayleigh(img_path)
    elif preprocess == 'RGHS':
        sceneRadiance = RGHS(img_path)
    elif preprocess == 'UCM':
        sceneRadiance = UCM(img_path)

    return sceneRadiance
```

```python
img_path = 'img.png'


image = cv2.imread('img.png')


CLAHE = image_enhancement(img_path, 'CLAHE')
GC = image_enhancement(img_path, 'GC')
HE = image_enhancement(img_path, 'HE')
ICM = image_enhancement(img_path, 'ICM')
Rayleigh = image_enhancement(img_path, 'Rayleigh')
RGHS = image_enhancement(img_path, 'RGHS')
UCM = image_enhancement(img_path, 'UCM')



# display results
fig, ax = plt.subplots(nrows=2, ncols=4, figsize=(10, 8))

ax = ax.flatten()
for a in ax:
    ax[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    ax[0].set_title('Original', fontsize=12)
    ax[0].set_xticks([])
    ax[0].set_yticks([])

    ax[1].imshow(CLAHE)
    ax[1].set_title('CLAHE', fontsize=12)
    ax[1].set_xticks([])
    ax[1].set_yticks([])

    ax[2].imshow(GC)
    ax[2].set_title('GC', fontsize=12)
    ax[2].set_xticks([])
    ax[2].set_yticks([])
    
    ax[3].imshow(HE)
    ax[3].set_title('HE', fontsize=12)
    ax[3].set_xticks([])
    ax[3].set_yticks([])
    
    ax[4].imshow(ICM)
    ax[4].set_title('ICM', fontsize=12)
    ax[4].set_xticks([]) 
    ax[4].set_yticks([]) 

    ax[5].imshow(Rayleigh)
    ax[5].set_title('Rayleigh', fontsize=12)
    ax[5].set_xticks([]) 
    ax[5].set_yticks([]) 

    ax[6].imshow(RGHS)
    ax[6].set_title('RGHS', fontsize=12)
    ax[6].set_xticks([]) 
    ax[6].set_yticks([]) 

    ax[7].imshow(UCM)
    ax[7].set_title('UCM', fontsize=12)
    ax[7].set_xticks([])
    ax[7].set_yticks([])
```
